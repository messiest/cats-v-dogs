```python
import os

import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import PIL
from IPython import display
```

```python
%matplotlib inline
```

```python
def load_image(file_path):
    img = PIL.Image.open(file_path).resize((150, 150))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    img = np.array(img)
    img = img / 255.
    img = np.expand_dims(img, axis=0)
    
    return img
```

```python
def predict_label(file_path, model):
    img = load_image(file_path)
    y = model.predict(img)
    if y > .5:
        return 'Dog'
    else:
        return 'Cat'
```

```python
saved_model = os.path.join('..', 'assets', 'models', 'cats-v-dogs', 'c25e161f-87ec-4600-910f-af79acc6ee66-model.h5')
```

```python
model = keras.models.load_model(saved_model)
```

```python
model.summary()
```

```python
predict_label('sample_images/cat.jpg', model)
```

```python
predict_label('sample_images/dog.jpeg', model)
```
