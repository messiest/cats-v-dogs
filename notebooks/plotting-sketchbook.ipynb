```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='darkgrid')
```

```python
!ls ../assets/results
```

```python
data = os.path.join('..', 'assets', 'results', 'cats_v_dogs.csv')
```

```python
df = pd.read_csv(data)
```

```python
df['epoch'] += 1
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 10), dpi=128)

plt.suptitle('Cats v. Dogs', fontsize=20)

sns.lineplot(x='epoch', y='acc', data=df, color='tab:orange', label="Training Accuracy", ax=ax1)
sns.lineplot(x='epoch', y='val_acc', data=df, color='tab:blue', label="Validation Accuracy", ax=ax1)
ax1.set_xlim(1, 15)
# ax1.set_ylim(0, 1)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')

sns.lineplot(x='epoch', y='loss', data=df, color='tab:orange', label="Training Loss", ax=ax2)
sns.lineplot(x='epoch', y='val_loss', data=df, color='tab:blue', label="Validation Loss", ax=ax2)
ax2.set_xlim(1, 15)
# ax2.set_ylim(0, 1)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')

plt.savefig('../assets/documentation/cats-v-dogs.png')
```
